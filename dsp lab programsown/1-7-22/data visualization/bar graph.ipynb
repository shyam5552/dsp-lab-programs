```python
import matplotlib.pyplot as plt
import numpy as np
country=["USA","INDIA","SINGAPORE","UK","JAPAN","CHINA"]
GDP=[16000,26000,19000,8000,12000,20021]
plt.title("BAR GRAPH")
plt.xlabel("country")
plt.ylabel("GDP")
plt.bar(country,GDP)
plt.show()
```

```python
colors=["red","blue","green","red","blue","red"]
plt.barh(country,GDP,color=colors)
plt.show()
```

```python
numbers=[0,2,4,3,5,6,7,8,5]
plt.hist(numbers,bins=3)
plt.show()
```

```python
y=np.array([20,40,35])
color=["r","b","green"]
plt.pie(y,colors=color,labels=["dsp","wt","or"])
plt.legend()
plt.show()
```
